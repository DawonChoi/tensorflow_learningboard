```python
import seaborn as sns
import matplotlib.pyplot as plt

import platform
from matplotlib import font_manager, rc
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['axes.unicode_minus'] = False

#윈도우
path = "c:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=path).get_name()
rc('font', family=font_name)

plt.barh(df['category'],df['vote'], color='red')
plt.title('제안분야별 투표수')
plt.xlabel('투표수')
plt.ylabel('제안분야')
plt.grid()

plt.show()
```

```python
df2 = df.groupby(by='category')[['vote']].sum()


df2 = df2.sort_values(by='vote', ascending=False)
```

```python
df2['vote'].sort_values().plot(kind='barh',color='red',label='vote')
plt.title('제안분야별 투표수')
plt.xlabel('투표수')
plt.ylabel('제안분야')
plt.grid()
plt.legend()

for i in range(3):
    plt.text(df2['vote'][i], 4-i, df2['vote'][i], fontsize=15)

#plt.savefig('제안분야별 투표수.png')
```
