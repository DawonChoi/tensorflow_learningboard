```python
import pandas as pd
CCTV_Seoul = pd.read_csv('../data/01. CCTV_in_Seoul.csv', encoding='utf-8')
CCTV_Seoul.head()
```

```python
c = CCTV_Seoul
c.rename(columns={c.columns[0]:'구별'}, inplace=True)
c.head()
```

```python
c.sort_values(by='소계', ascending=True).head()
```

```python
c.sort_values(by='소계', ascending=False).head()
```

```python
c['최근증가율']=(c['2016년']+c['2015년']+c['2014년'])/c['2013년도 이전']*100
c.sort_values(by="최근증가율", ascending=False).head()
```

```python
p = pd.read_excel('../data/01. population_in_Seoul.xls', header=2, encoding='utf-8')
p.head()
```

```python
p = pd.read_excel('../data/01. population_in_Seoul.xls', header=2, usecols ='B, D, G ,J, N',encoding='utf-8')
p.head()
```

```python
p.rename(columns={p.columns[0]:'구별', p.columns[1]:'인구수', p.columns[2]:'한국인', p.columns[3]:'외국인', p.columns[4]:'고령자'}, inplace=True)
p.head()
```

```python
p.drop([0], inplace=True)
p.head()
```

```python
p['구별'].unique()
```

```python
p[p['구별'].isnull()]
```

```python
p.drop([26], inplace=True)
p.head()
```

```python
p['외국인 비율']= p['외국인'] / p['인구수'] * 100
p['고령자 비율']= p['고령자'] / p['인구수'] * 100
p.head()
```

```python
p.sort_values(by='인구수', ascending=False).head()
```

```python
p.sort_values(by='한국인', ascending=False).head()
```

```python
p.sort_values(by='외국인', ascending=False).head()
```

```python
p.sort_values(by='외국인 비율', ascending=False).head()
```

```python
p.sort_values(by='고령자 비율', ascending=False).head()
```

```python
result = pd.merge(c,p, on='구별')
result.head()
```

```python
del result['2013년도 이전']
del result['2014년']
del result['2015년']
del result['2016년']
result.head()
```

```python
result.set_index('구별', inplace=True)
result.head()
```

```python
corr = result.corr(method = 'pearson')
corr
```

```python
import statsmodels.api as sm
result_model = sm.OLS.from_formula("소계 ~ 인구수", result).fit()
result_model.summary() 
```

```python
result['CCTV비율'] = result['소계'] / result['인구수'] * 100
result['CCTV비율'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(result['인구수'], result['소계'], s=50)
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
import numpy as np

fp1 = np.polyfit(result['인구수'], result['소계'], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(result['인구수'], result['소계'], s=50)
plt.plot(fx, f1(fx), linestyle='dashed', lw=3, color='g')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
result['오차'] = np.abs(result['소계'] - f1(result['인구수']))
df_sort = result.sort_values(by='오차', ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(result['인구수'], result['소계'], c=result['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(10):
    plt.text(df_sort['인구수'][n]*1.02, df_sort['소계'][n]*0.98, df_sort.index[n], fontsize=15)
    
plt.xlabel('인구수')
plt.ylabel('인구당비율')

plt.colorbar()
plt.grid()
plt.show()
```
